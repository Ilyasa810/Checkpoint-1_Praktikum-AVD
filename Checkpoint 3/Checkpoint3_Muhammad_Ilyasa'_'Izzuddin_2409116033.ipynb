```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/euro2024_players.csv")
df
```

## Blox Plot Distribusi Caps Berdasarkan Posisi Pemain

```python
# Boxplot (Seaborn)
plt.figure(figsize=(12, 7))
sns.boxplot(x='Position', y='Caps', data=df)
plt.title('Distribusi Caps Berdasarkan Posisi Pemain')
plt.xlabel('Posisi Pemain')
plt.ylabel('Caps')
plt.xticks(rotation=30)
plt.show()
```

### Insight Distribusi Caps Berdasarkan posisi pemain
Distribusi *caps* bervariasi di tiap posisi, dengan *Centre-Back*, *Central Midfield*, dan *Centre-Forward* memiliki rentang yang luas dan banyak *outliers*, sementara *Goalkeeper* dan *Left Midfield* cenderung lebih rendah. Posisi *Second Striker* memiliki distribusi *caps* yang lebih terkonsentrasi, sedangkan *Right Winger* dan *Second Striker* memiliki median yang lebih tinggi dibanding posisi lain.


### Tindakan yang dapat diambil selanjutnya
Untuk menjaga keseimbangan tim, rekrutmen atau pengembangan pemain dapat difokuskan pada posisi dengan *caps* rendah, sementara rotasi lebih sering diperlukan di posisi dengan variasi *caps* kecil. Selain itu, strategi regenerasi dapat diterapkan untuk mengatasi ketimpangan antara pemain veteran dan pemain baru.

## Histogram Jumlah Penampilan Pemain di Tim Nasional

```python
# Histogram (Seaborn)
plt.figure(figsize=(8, 5))
sns.histplot(df['Caps'], bins=20, kde=True, color='navy')
plt.title('Jumlah Penampilan Pemain di Tim Nasional')
plt.xlabel('Jumlah Penampilan')
plt.ylabel('Pemain')
plt.show()
```

### Insight Jumlah Penampilan pemain di Tim Nasional

Berdasarkan grafik diatas dapat dilihat bahwa, banyak pemain yang belum pernah tampil atau dipanggil tim nasional sebelumnya sedangkan untuk pemain dengan jumlah penampilan diatas 100 penampilan lebih sedikit. Faktor yang dapat mempengaruhi hal tersebut adalah minimnya jam terbang bermain di klub sehingga membuat pelatih tim nasional tidak tertarik memanggil pemain yang minim jam terbang bermain di klub tersebut. Dan juga, umur yang masih muda bisa menjadi salah satu faktor mengapa banyak pemain yang belum pernah dipanggil untuk mewakili negaranya berlaga.


### Tindakan yang dapat dilakukan selanjutnya

Pelatih tim nasional dapat mencari atau mencoba pemain-pemain muda atau pemain yang bermain bagus di klubnya untuk dipanggil ke tim nasionalnya, karena sebuah tim nasional dapat dikatakan sukses dalam hal regenerasi ketika tim nasional tersebut dapat beregenerasi skuad dengan baik. Seperti contohnya timnas spanyol, mereka banyak memanggil pemain muda dan pemain yang belum pernah dipanggil timnas sebelumnya seperti Lamine Yamal, Alex Baena, Fabian Ruiz, Nico Williams dll. Dan mereka dapat bersaing dikompetisi tersebut. Dan dengan dipanggilnya ke tim nasional dan mereka bermain bagus itu dapat menaikkan value mereka, sehingga dilirik oleh klub-klub besar.

## Bar Chart Jumlah Pemain Berdasarkan Posisi

```python
# Bar Chart (Seaborn)
plt.figure(figsize=(10, 6))
sns.countplot(x='Position', data=df)
plt.title('Jumlah Pemain Berdasarkan Posisi')
plt.xlabel('Posisi Pemain')
plt.ylabel('Jumlah Pemain')
plt.xticks(rotation=30)
plt.show()
```

### Insight dari Jumlah Pemain berdasarkan posisi

Berdasarkan grafik tersebut dapat dilihat bahwa, banyak pemain di posisi *Centre-Back* yang dipanggil atau dibawa tim nasional dan juga ada tiga posisi yang memiliki jumlah pemain paling sedikit diposisinya yaitu *Left Midfielder*, *Right Midfielder*, dan *Second Striker*.

### Tindakan yang dapat dilakukan selanjutnya

Pelatih tim nasional dapat melakukan pencarian pemain lebih merata dari berbagai posisi dan tidak lupa menyesuaikan kebutuhan dan kedalaman skuad dari tim nasional tersebut.

## Pie Chart Distribusi Pemain Berdasarkan Negara

```python
# Pie Chart (Matplotlib)
plt.figure(figsize=(8, 8))
country_counts = df['Country'].value_counts()
country_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['red', 'silver', 'lightgreen', 'gold'])
plt.title('Distribusi Pemain Berdasarkan Negara')
plt.ylabel('')
plt.show()
```

### Insight Distribusi Pemain Berdasarkan Negara

Terlihat bahwa tidak ada negara yang mendominasi secara signifikan, dengan sebagian besar negara memiliki proporsi pemain yang hampir sama di kisaran 4.2%. Dengan Tim memiliki yang keberagaman yang tinggi dan pemain dari banyak negara, dapat memberikan variasi dalam gaya bermain dan pengalaman internasional. Dengan banyaknya asal negara yang berbeda, mungkin ada tantangan dalam komunikasi dan adaptasi gaya bermain tim.

### Tindakan yang dapat dilakukan selanjutnya

Untuk meningkatkan kohesi tim dan efektivitas permainan, dapat dilakukan program pelatihan berbasis komunikasi, analisis strategi berdasarkan gaya bermain masing-masing negara, serta dukungan adaptasi seperti kursus bahasa agar pemain lebih cepat berintegrasi.

## Bar Plot Top 20 Pencetak Gol Terbanyak

```python
# Bar Plot (Seaborn)
plt.figure(figsize=(9, 6))
top_10_goals = df[['Name', 'Goals']].sort_values(by='Goals', ascending=False).head(20)
sns.barplot(x='Goals', y='Name', data=top_10_goals, palette='viridis')
plt.title('Top 20 Pencetak Gol Terbanyak')
plt.xlabel('Gol')
plt.ylabel('Pemain')
plt.show()
```

### Insight Top 20 Pencetak Gol Terbanyak

Berdasarkan grafik tersebut, Cristiano Ronaldo menjadi pencetak gol terbanyak di kompetisi tersebut dengan perolehan gol lebih dari 120 gol, disusul Romelu Lukaku dan Robert Lewandowski di urutan kedua dan ketiga.

### Tindakan yang dapat dilakukan selanjutnya

Untuk meningkatkan performa tim, dapat dilakukan analisis pola permainan pencetak gol terbanyak, strategi rekrutmen atau rotasi pemain, evaluasi kontribusi di berbagai kompetisi, serta optimalisasi peran playmaker untuk meningkatkan peluang mencetak gol.



## Bar Plot Top 10 Pemain dengan Market Value Tertinggi

```python
# Bar Plot (Seaborn)
plt.figure(figsize=(10, 6))
top_10_valued_players = df[['Name', 'MarketValue']].sort_values(by='MarketValue', ascending=False).head(10)
sns.barplot(x='MarketValue', y='Name', data=top_10_valued_players, palette='viridis')
plt.title('Top 10 Pemain dengan Market Value Tertinggi')
plt.xlabel('Nilai Pasar (juta EUR)')
plt.ylabel('Pemain')
plt.show()
```

### Insight Top 10 Pemain dengan Market Value Tertinggi

Jude Bellingham menjadi pemain dengan market value tertinggi dengan harga diatas 1,75 juta Euro atau sekitar 30 miliar rupiah, disusul Kylian Mbappe dan Phil Foden diposisi dua dan tiga.

### Tindakan yang dapat dilakukan selanjutnya
Pemain muda dapat memiliki market value atau nilai pasar yang tinggi karena memiliki waktu bermain yang tinggi pula. Sehingga waktu bermain yang tinggi bisa menjadi salah satu faktor tingginya market value sebuah pemain.

## Bar Plot Top 20 Pemain dengan Caps Terbanyak Berdasarkan Posisi

```python
# Bar Plot (Seaborn)
plt.figure(figsize=(10, 6))
top_caps = df[['Name', 'Caps', 'Position']].sort_values(by='Caps', ascending=False).head(20)
sns.barplot(x='Caps', y='Name', hue='Position', data=top_caps, palette='Set2')
plt.title('Top 20 Pemain dengan Caps Terbanyak Berdasarkan Posisi')
plt.xlabel('Caps')
plt.ylabel('Pemain')
plt.show()
```

### Insight Top 20 Pemain dengan Caps terbanyak

Pemain dengan caps tertinggi didominasi oleh Cristiano Ronaldo, Luka Modric, dan Robert Lewandowski, dengan posisi Centre-Forward, Central Midfield, dan Centre-Back yang cenderung memiliki caps lebih tinggi, menunjukkan bahwa pengalaman dan konsistensi memainkan peran penting dalam menjaga performa di level tertinggi.

### Tindakan yang dapat dilakukan selanjutnya

Untuk meningkatkan kualitas tim, analisis lebih lanjut bisa dilakukan terhadap faktor-faktor yang mendukung pemain dengan caps tinggi agar bisa diaplikasikan ke pemain lain, melakukan regenerasi pada posisi dengan caps tinggi untuk menghindari ketergantungan pada pemain senior, serta mengembangkan strategi rotasi yang lebih efektif agar distribusi caps lebih merata di seluruh posisi.
